```python
word = 'hockey'

Chances = 10

Empty = []


print('Welcome to Hangman game! ')
print('Hint : Sports')

number_of_underscores = len(word)


for i in range(number_of_underscores):
    Empty.append('_')
    update = ' '.join(Empty)
print(update)
        
while Chances > 0:
    guess_letter = input('Guess a letter: ').lower()

    if guess_letter in word:
        print('Remaining Chances: ',Chances)
        print('  Matched  ')
    
        position = word.find(guess_letter)
      
        Empty[position] = guess_letter
    
        update = ' '.join(Empty)
        print('\n',update)

    
    else:
        print()
        print('  Wrong Guess  ')
    
        Chances = Chances-1
    
        print('Remaining chances: ',Chances)
    
        print('\n',update)
        
    if Chances == 0:
        print('Game lost! ')
    elif '_' not in Empty:
        print('\nCongrats! You won the game!')
        break
    
```
